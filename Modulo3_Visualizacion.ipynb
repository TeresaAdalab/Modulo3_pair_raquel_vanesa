#### 4. ¿Cuál es la mediana de la esperanza de vida en Europa? Proporciona un rango intercuartil aproximado y verifica si existen valores atípicos.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv('df_vf_modificado.csv')

# Filtrar datos solo para países europeos (suponiendo que la columna 'abbreviation' indica el país)
# Esto puede variar según la estructura exacta del dataset
df_europe = df[df['abbreviation'].isin(['AL', 'AD', 'AT', 'BY', 'BE', 'BA', 'BG', 'HR', 'CY', 'CZ', 'DK', 'EE', 'FI', 'FR', 'DE', 'GR', 'HU', 'IS', 'IE', 'IT', 'LV', 'LI', 'LT', 'LU', 'MT', 'MD', 'MC', 'ME', 'NL', 'MK', 'NO', 'PL', 'PT', 'RO', 'RU', 'SM', 'RS', 'SK', 'SI', 'ES', 'SE', 'CH', 'UA', 'GB'])]

# Calcular la mediana de la esperanza de vida en Europa
median_life_expectancy = df_europe['lifeexpectancy'].median()

# Calcular el rango intercuartil (IQR) para la esperanza de vida en Europa
Q1 = df_europe['lifeexpectancy'].quantile(0.25)
Q2 = df_europe['lifeexpectancy'].quantile(0.50)
Q3 = df_europe['lifeexpectancy'].quantile(0.75)
Q4 = df_europe['lifeexpectancy'].quantile(1.00)
IQR = Q4 - Q1

# Calcular límites para identificar valores atípicos
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q4 + 1.5 * IQR

# Identificar valores atípicos
outliers = df_europe[(df_europe['lifeexpectancy'] < lower_bound) | (df_europe['lifeexpectancy'] > upper_bound)]

# Imprimir resultados
print(f"Mediana de esperanza de vida en Europa: {median_life_expectancy}")
print(f"Rango intercuartil (IQR) aproximado: {IQR}")
print(f"Valores atípicos identificados:\n{outliers[['abbreviation', 'lifeexpectancy']]}")

# Visualización utilizando boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='lifeexpectancy', data=df_europe)
plt.title('Distribución de la esperanza de vida en Europa')
plt.xlabel('Esperanza de vida')
plt.show()
```

#### 6. ¿Cuáles son los 10 países con la mayor área forestal?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el DataFrame
df = pd.read_csv('df_vf_modificado.csv')

# Ordenar por área forestal en orden descendente y seleccionar los top 10
top_forest_countries = df.sort_values(by='forestedarea(%)', ascending=False).head(10)

# Seleccionar los 10 países con mayor área forestal
top_10_forest_countries = top_forest_countries.head(10)

# Mostrar los resultados
print(top_10_forest_countries[['abbreviation', 'forestedarea(%)']])

# Configurar el gráfico de barras horizontales
plt.figure(figsize=(10, 6))
sns.barplot(x='forestedarea(%)', y='abbreviation', data=top_forest_countries, palette='viridis')
plt.xlabel('Área forestal (%)')
plt.ylabel('Países')
plt.title('Top 10 Países con Mayor Área Forestal')
plt.xlim(0, top_forest_countries['forestedarea(%)'].max() * 1.1)  # Ajustar límite del eje x
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Agregar líneas de grid en el eje x
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el DataFrame
df = pd.read_csv('df_vf_modificado.csv')

# Ordenar por área forestal en orden descendente y seleccionar los top 10
top_forest_countries = df.sort_values(by='forestedarea(%)', ascending=False).head(10)

# Preparar datos para el gráfico de pastel
labels = top_forest_countries['abbreviation']
sizes = top_forest_countries['forestedarea(%)']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
          '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

# Configurar el gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Top 10 Países con Mayor Área Forestal')
plt.axis('equal')  # Hace que el gráfico de pastel sea un círculo
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el DataFrame
df = pd.read_csv('df_vf_modificado.csv')

# Seleccionar las columnas de interés
columns_of_interest = ['birthrate', 'forestedarea(%)']

# Filtrar el DataFrame para asegurar que no haya valores nulos o NaNs
df_filtered = df[columns_of_interest].dropna()

# Configurar el estilo de seaborn
sns.set(style="whitegrid")

# Crear el gráfico combinado de histograma y regresión
plt.figure(figsize=(10, 6))
sns.regplot(x='forestedarea(%)', y='birthrate', data=df_filtered, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title('Relación entre Área Forestal y Tasa de Natalidad')
plt.xlabel('Área Forestal (%)')
plt.ylabel('Tasa de Natalidad')
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
