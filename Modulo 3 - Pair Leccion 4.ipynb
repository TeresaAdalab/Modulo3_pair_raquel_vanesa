# B) Ejercicios de Limpieza

Después de la unión de datos, tenemos dos columnas de "country". Elimina una de ellas.

```python

#2.Los nombres de las columnas no son homogeneos.
## Espacios vacios
df_vf.rename(columns=lambda x: x.replace(' ', ''), inplace=True)
```

```python
df_vf.rename(columns=lambda x: x.replace(' ', '').replace('\n', '').replace(':', '').replace(r'\(.*\)', '').lower(), inplace=True)
print(df_vf.head())
```

```python
##3·La columnas  tiene la latitud y la longitud en una sola columna. 
#Dividir la columna "coordinates" en dos columnas separadas de longitud y latit
df_vf[['longitude', 'latitude']] = df_vf['coordinates'].str.split(',', expand=True)
print(df_vf.head())

```

```python
print(df_vf.columns)
```

```python
# Eliminar  la columna "coordinates" 
df_vf.drop(columns=['coordinates'], inplace=True)
```

```python
print(df_vf.columns)
```

```python
##4·Elimina los "%" de los valores de las columnas 

columnas = ['unemploymentrate', 'totaltaxrate', 'taxrevenue(%)', 'populationlaborforceparticipation(%)', 'outofpockethealthexpenditure', 'grosstertiaryeducationenrollment(%)', 'grossprimaryeducationenrollment(%)', 'forestedarea(%)', 'cpichange(%)', 'agriculturalland(%)']


df_vf[columnas] = df_vf[columnas].replace('%', '', regex=True)
```

```python
df_vf.head(2)
```

```python
##5. Eliminar $ Columnas de gasoline_price, gdp, minimum_wage, pero eliminando "$".
```

```python
# Columnas que contienen el símbolo "$"
columnas_dollar = ['gasolineprice', 'gdp', 'minimumwage']

# Eliminar el símbolo "$" de los valores en las columnas especificadas
df_vf[columnas_dollar] = df_vf[columnas_dollar].replace('$', '', regex=False)

# Mostrar información estadística sobre las columnas seleccionadas
print(df_vf[columnas_dollar].describe())
```

```python
df_vf.to_csv('df_vf_modificado.csv', index=False)
```

# C) Ejercicios de Filtrado

```python
#1. Países cuya mortalidad infantil esté entre 40 y 50 personas por kilómetro cuadrado.
paises_mortalidad_infantil = df_vf[(df_vf['infantmortality'] >= 40) & (df_vf['infantmortality'] <= 50)]
print(paises_mortalidad_infantil)
```

```python

#2.Encuentra los países cuyas tasas de natalidad son mayores o iguales a 20 y su esperanza de vida es mayor de 75 años.

paises_seleccionados = df_vf[(df_vf['birthrate'] >= 20) & (df_vf['lifeexpectancy'] > 75)]

# Mostrar los países seleccionados
print(paises_seleccionados)
```

```python
#3. Encuentra las ciudades cuyos paises contienen la palabra "la" en su nombre.
```

```python

# Seleccionar las ciudades cuyos países contienen la palabra "la" en su nombre
df_vf_cleaned = df_vf.dropna(subset=['largestcity']) #eliminamos los NA de largestcity
ciudades_seleccionadas = df_vf_cleaned[df_vf_cleaned['largestcity'].str.contains('la', case=False)]
print(ciudades_seleccionadas['largestcity'])
```

```python
#4. países cuyos medicos por cada 1000 habitantes (physicians_per_thousand) sea mayores de 5.
paises_seleccionados = df_vf[df_vf['physiciansperthousand'] > 5]

print("\n Paises con mas de medicos :\n")
print(paises_seleccionados['physiciansperthousand'])
```

```python

#5.Encuentra los países cuyatasa de fertilidad sea mayor a 6.
paises_fertilidad = df_vf[df_vf['fertilityrate'] > 6]
paises_fertilidad
print(paises_fertilidad['fertilityrate'])
```

```python
#6. Países cuya moneda es el euro (EUR) y tienen una tasa de natalidad superior al promedio.
```

```python
promedio_natalidad = df_vf['birthrate'].mean()

paises_euro = df_vf[df_vf['currency-code'] == 'EUR']

paises_seleccionados = paises_euro[paises_euro['birthrate'] > promedio_natalidad]
print(paises_seleccionados[['birthrate', 'currency-code']])

```

```python
# Seleccionar los países cuyas tasas de mortalidad infantil son superiores a 70
paises_seleccionados = df_vf[df_vf['infantmortality'] > 70]

# Mostrar los países seleccionados
print(paises_seleccionados[ 'infantmortality'])
```
