### ESTADÍSTICA DESCRIPTIVA

```python
import pandas as pd

# Cargar el DataFrame
df = pd.read_csv('df_vf_modificado.csv')
```


1. Calcula las siguientes estadísticas descriptivas para las variables numéricas:

- Media

- Mediana

- Desviación estándar

- Mínimo

- Máximo

```python

# 1. Calcular estadísticas descriptivas

import matplotlib.pyplot as plt

# Variables numéricas
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Estadísticas descriptivas
mean = df[numeric_columns].mean()
median = df[numeric_columns].median()
std_dev = df[numeric_columns].std()
min_val = df[numeric_columns].min()
max_val = df[numeric_columns].max()

print("\nMedia:\n", mean)
print("\nMediana:\n", median)
print("\nDesviación Estándar:\n", std_dev)
print("\nMínimo:\n", min_val)
print("\nMáximo:\n", max_val)

numeric_data = df.select_dtypes(include='number')
# Crear un boxplot para cada variable numérica
for column in numeric_data.columns:
    plt.figure(figsize=(8, 6))
    numeric_data.boxplot(column=column)
    plt.title(f'Boxplot de {column}')
    plt.xlabel('Variable')
    plt.ylabel('Valores')
    plt.show()
```

4. Realiza un análisis de la mortalidad infantil (infant_mortality) y clasifica los países en tres categorías: baja, moderada y alta mortalidad infantil.

```python

# 4. Análisis de la mortalidad infantil (infant_mortality)
def infant_mortality(row):
    if row['infantmortality'] < 10:
        return 'Baja'
    elif row['infantmortality'] < 30:
        return 'Moderada'
    else:
        return 'Alta'

df['infantmortalityclass'] = df.apply(infant_mortality, axis=1)
infant_mortality_counts = df['infantmortalityclass'].value_counts()

print("\nClasificación de mortalidad infantil:\n") 
print(infant_mortality_counts)

percentiles = df['infantmortality'].quantile([0.33, 0.66])

# Función para clasificar la mortalidad infantil en categorías
def classify_infant_mortality(mortality_rate):
    if mortality_rate < percentiles[0.33]:
        return 'Baja'
    elif mortality_rate < percentiles[0.66]:
        return 'Moderada'
    else:
        return 'Alta'
# Aplicar la función para crear una nueva columna 'mortality_category'
df_vf['mortalitycategory'] = df_vf['infantmortality'].apply(classify_infant_mortality)
# Mostrar los resultados
print("Clasificación de la mortalidad infantil por país:")
print(df_vf[['abbreviation', 'infantmortality', 'mortalitycategory']])
```

5. Calcula la correlación entre las variables de "fertility_rate" y "life_expectancy". Interpreta la relación entre estas dos variables. ¿Qué tipo de correlación hay? Explica la razón de esta relación

```python

# 5. Calcular la correlación entre 'fertilityrate' y 'lifeexpectancy'
correlation = df['fertilityrate'].corr(df['lifeexpectancy'])
print(f"\nCorrelación entre 'fertilityrate' y 'lifeexpectancy': {correlation}")

# Interpretación de la correlación
if correlation > 0:
    relation_type = "positiva"
else:
    relation_type = "negativa"

print(f"\nLa relación entre la tasa de fertilidad y la esperanza de vida es {relation_type}. Esto puede deberse a que en países con mayor fertilidad, los recursos por niño son menores, afectando negativamente a la esperanza de vida.")
```

9. Genera un mapa de calor (heatmap) de la matriz de correlación de todas las variables numéricas y discute las relaciones más destacadas.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Seleccionar solo las columnas numéricas para la matriz de correlación
numeric_columns = df.select_dtypes(include=np.number)
# Calcular la matriz de correlación
correlation_matrix = numeric_columns.corr()
# Crear una máscara para la mitad superior del mapa de calor
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
# Configurar el tamaño del gráfico
plt.figure(figsize=(20, 15))
# Generar el mapa de calor de la matriz de correlación usando solo la mitad superior
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5,mask=mask)
# Añadir título
plt.title('Mapa de Calor de la Mitad Superior de la Matriz de Correlación')
# Mostrar el mapa de calor
plt.show()
```
